```python
%matplotlib inline

from numpy import*
from matplotlib.pyplot import*
```

```python
r=linspace(4,10,1000)

omegamax=(-r*sqrt(r**2-2*r+0.25)+1)/(-r**3+0.25*r+0.5)
omegamin=(r*sqrt(r**2-2*r+0.25)+1)/(-r**3+0.25*r+0.5)

omega1=(-r*sqrt(r**2-2*r+0.25)+0.5*r+0.5)/(-r**4+2*r**3+0.5*r+0.25)
omega2=(r*sqrt(r**2-2*r+0.25)+0.5*r+0.5)/(-r**4+2*r**3+0.5*r+0.25)

omegacosta=0.5/(r**2+0.25)
```

```python
fig1, ax=subplots()

ax.plot(r,omegamax,label=r'$\omega_{\mathrm{lim}}$',color='red')
ax.plot(r,omegamin,color='red')
ax.plot(r,omega1,label=r'$\omega_1$',color='green')
ax.plot(r,omega2,label=r'$\omega_2$',color='blue')
ax.plot(r,omegacosta,label=r'$\omega_{\mathrm{Costa}}$',color='black')

ax.set_title(r'$\omega$ vs $r$')
ax.set_xlabel(r'$r$')
ax.set_ylabel(r'$\omega$')
ax.set_xlim([4,10])
ax.legend()

fig1.savefig('grafico1.pdf')
```
